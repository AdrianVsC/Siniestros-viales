```python
import pandas as pd
```

```python
archivo_homi = r"C:\Users\argui\OneDrive\Escritorio\Siniestros viales\Dataset\homicidiosETL.csv"
archivo_vic = r"C:\Users\argui\OneDrive\Escritorio\Siniestros viales\Dataset\victimasETL.csv"
```

Insertamos los datos en un DataFrame

```python
#Insertamos Homicidios
df_h = pd.read_csv(archivo_homi,delimiter=';',encoding='latin-1',index_col=0)
df_h.isna().sum()
```

```python
#Insertamos Victimas
df_v = pd.read_csv(archivo_vic,delimiter=';',encoding='latin-1',index_col=0)
df_v.isna().sum()
```

```python
#Debido a que Homicidios ya tiene fecha, eliminamos fecha de victimas
df_v = df_v.drop(columns='Fecha')
```

Realizamos el merge de las tablas para mejorar el análisis

```python
df_final = pd.merge(df_v,df_h,on='Id')
df_final.head(3)
```

Analizamos los datos del merge

```python
df_final.isna().sum()
#Observamos que las columnas que tienen nulos son las de tipo datetime.
#El nulo NaT es manejable con pandas 
```

```python
'''
Analizamos las columnas, observamos que tenemos 2 fechas Victima_x y Victima_y
por ello verificaremos si tienen datos iguales o similares y procederemos a eliminar 1 de ellas
'''
iguales = (df_final['Victima_x'] == df_final['Victima_y']).all()
if iguales:
          print('Son iguales')
else: print('No son iguales y las columnas son: ')
df_final['diferente'] = df_final['Victima_x'] != df_final['Victima_y']
df_final[df_final['diferente']==True]

#Observo que la columna que da mejor información es Victima_X
```

```python
#Eliminamos la columna que creamos
df_final = df_final.drop(columns='diferente')

#Usaremos Fecha_x
df_final = df_final.drop(columns='Victima_y')
df_final = df_final.rename(columns={'Victima_x':'Victima'})
df_final.head(3)
```
