```python
#Importamos las librerías

import pandas as pd
```

Insertamos los datos en un DataFrame para su posterior análisis


```python
#inspeccionamos los datos
df_homi= pd.read_csv(r"C:\Users\argui\OneDrive\Escritorio\Siniestros viales\Dataset\homicidios.csv",delimiter=';',encoding='latin-1')
df_homi.head()
```

Analizamos y limpiamos el contenido del Dataframe

```python
df_homi.isna().sum()
#Observamos que las columnas tienen muchas filas vacías 
```

```python
#Eliminamos las filas que tengan todas las columbas vacías 
df_homi_2 = df_homi.dropna(axis=0, how='all')
df_homi_2.isna().sum() #Reducimos mucho la cantidad de nulos
```

```python
#Verificamos si hay columnas con todas las columnas repetidas
duplicados = df_homi_2[df_homi_2.duplicated(keep=False)]
duplicados
#No hay filas repeditas
```

Asignamos nombres que representen de mejor manera el contenido de la columna 

```python
#Primero normalizamos el nombre de las columnas

# Función para normalizar el nombre de la columna
def normalizar_nombre_columna(nombre):
    # Dividir el nombre por espacios
    partes = nombre.split()
    # Si hay más de una parte (es decir, había espacios), unir con guion bajo y capitalizar cada palabra
    if len(partes) > 1:
        nombre_normalizado = '_'.join([parte.capitalize() for parte in partes])
    else:
        # Si no hay espacios, capitalizar la primera letra
        nombre_normalizado = nombre.capitalize()
    return nombre_normalizado

# Normalizar nombres de columnas en el DataFrame
df_homi_2.columns = [normalizar_nombre_columna(col) for col in df_homi_2.columns]

# Mostramos los nombres de las columnas normalizados

df_homi_2.head(1)
```

```python
#Ahora cambiamos asignamos los nombres que representen mejor los datos
df_homi_2 = df_homi_2.rename(columns={'Pos_X':'Latitud', 'Pos_Y':'Longitud'})
df_homi_2 = df_homi_2.rename(columns={'Dirección_Normalizada':'Dirección'})
df_homi_2 = df_homi_2.rename(columns={'Hora':'Hora_no'})
df_homi_2 = df_homi_2.rename(columns={'Hh':'Hora'})
df_homi_2 = df_homi_2.rename(columns={'Aaaa':'Año','Mm':'Mes','Dd':'Dia'})
df_homi_2.head(1)
```

Eliminamos las columnas que no nos será útil

```python
#Eliminaremos las columnas que separan la fecha y la hora ya que pandas tiene métodos con los cuales trabajaremos
df_homi_2 = df_homi_2.drop(columns=['Fecha'])
df_homi_2 = df_homi_2.drop(columns=['Hora_no'])
df_homi_2 = df_homi_2.drop(columns=['Xy_(caba)']) #Suficiente con las
df_homi_2 = df_homi_2.drop(columns=['Lugar_del_hecho','Participantes','Tipo_de_calle'] )
df_homi_2.head(1)
```

Cambiamos el tipo de datos de algunas columnas para facilitar el análisis

```python
#Para el análisis con pandas de las columnas, necesitamos que el tipo de datos sea el adecuado

#Primero nos aseguramos que los valores no numéricos sean convertidos a Nan para luego trabajarlos
df_homi_2['Hora'] = pd.to_numeric(df_homi_2['Hora'],errors='coerce')
df_homi_2['Mes'] = pd.to_numeric(df_homi_2['Mes'],errors='coerce')
df_homi_2['Dia'] = pd.to_numeric(df_homi_2['Dia'],errors='coerce')
df_homi_2['Comuna'] = pd.to_numeric(df_homi_2['Comuna'],errors='coerce')
df_homi_2['Hora'] = pd.to_numeric(df_homi_2['Hora'],errors='coerce')
```

Imputamos datos 

```python

#Imputamos SD donde la altura sea nulo
df_homi_2['Cruce'] = df_homi_2['Cruce'].fillna('SD')

#Imputamos Sd donde Calle sea nulo
df_homi_2['Calle'] = df_homi_2['Calle'].fillna('SD')

#Imputamos SD donde Dirección sea
df_homi_2['Dirección'] = df_homi_2['Dirección'].fillna('SD')

df_homi_2.isna().sum()

```

```python
'''Explorando los datos, me percaté que en Latitud y longitud, 
hay otra manera de rellenar los datos diferente a 'SD' y es colocando '.' '''

# Reemplaza las comas por puntos en las columnas de coordenadas si es necesario
df_homi_2['Longitud'] = df_homi_2['Longitud'].str.replace(',', '.')
df_homi_2['Latitud'] = df_homi_2['Latitud'].str.replace(',', '.')

# Identifica valores no convertibles
invalid_longitudes = df_homi_2[~df_homi_2['Longitud'].str.replace('.', '', 1).apply(lambda x: x.isdigit() or x.replace('-', '').isdigit())]
invalid_latitudes = df_homi_2[~df_homi_2['Latitud'].str.replace('.', '', 1).apply(lambda x: x.isdigit() or x.replace('-', '').isdigit())]

print("Valores inválidos en 'Longitud':")
print(invalid_longitudes)

print("Valores inválidos en 'Latitud':")
print(invalid_latitudes)

# Elimina o maneja los valores no convertibles
df_homi_2 = df_homi_2.drop(invalid_longitudes.index)
df_homi_2 = df_homi_2.drop(invalid_latitudes.index)

# Convierte las columnas a float
df_homi_2['Longitud'] = df_homi_2['Longitud'].astype(float)
df_homi_2['Latitud'] = df_homi_2['Latitud'].astype(float)

# Muestra el DataFrame resultante
df_homi_2.info()
```
